```python
import cv2
import matplotlib.pyplot as plt
import os.path
```

```python
path = 'C:\\Users\\kenta\\Desktop\\apex\\jissou\\image_rotate'
name = 'mato3'
types = '.png'
name_image = name + types
image_path = os.path.join(path,name_image)
img = cv2.imread(image_path)
```

```python
plt.imshow(img)
```

```python
def rotate(img):

    height = img.shape[0]                         
    width = img.shape[1]                         
    center = (int(width/2), int(height/2))
    for i in range(5):
        angle = i *45
        scale = 1.0
        trans = cv2.getRotationMatrix2D(center, angle , scale)
        image2 = cv2.warpAffine(img, trans, (width,height))
        plt.imshow(img)
        plt.imshow(image2)
        rotate_image_name = name + '_{}.png'.format(angle)
        #print(rotate_image_name)
        path2 = os.path.join(path,rotate_image_name)
        print(path2)
        cv2.imwrite(path2,image2)
rotate(img)
```

```python
#高さを定義
height = img.shape[0]                         
#幅を定義
width = img.shape[1]  
#回転の中心を指定                          
center = (int(width/2), int(height/2))
```

```python
#回転角を指定
angle = 180.0
#スケールを指定
scale = 1.0
#getRotationMatrix2D関数を使用
trans = cv2.getRotationMatrix2D(center, angle , scale)
#アフィン変換
image2 = cv2.warpAffine(img, trans, (width,height))
```
